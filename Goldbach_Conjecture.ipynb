```python
n=1000
```

```python
from sympy import *
def Goldbach(n):
    for i in range(n):
        for j in range(n):
            if isprime(i) and isprime(j):
                if n==i+j:
                    print(n,"=",i,"+",j)
count=0
for i in range(4,n+1,2):
    print("*"*15)
    Goldbach(i)
    count=count+1
    print("*"*15)
print("Total even integers for which Goldbach Conjecture is satisfied is: ",count)
# no of even integers between 4 and n
z=int((n-4)/2+1)
```

```python
if z==count:
    print("Goldbach Conjecture is satisfied")
else:
    print("Goldbach Conjecture is not satisfied")
```
